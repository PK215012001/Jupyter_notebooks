Plotting SRH vs RR score ipl

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
overs = [x for x in range(1, 21)]
srh_score = [10, 24, 45, 55, 78, 94, 106, 115, 123, 135, 147, 156, 178, 196, 208, 219, 233, 256, 273, 286]

def convert_over_format(over_list):
    return [int(o) + (o - int(o)) * 10 / 6 for o in over_list]

srh_wicket_overs = convert_over_format([3.1, 9.3, 14.2, 18.2, 19.2, 19.3])
srh_wicket_score = [45, 130, 202, 258, 279, 279]

plt.xticks(overs)
plt.plot(overs, srh_score, label = 'SRH', color = '#E65100', linewidth = 2.5)
plt.plot(srh_wicket_overs, srh_wicket_score, linestyle = 'None', marker = 'x', markersize = 12, markeredgewidth = 3, markeredgecolor = '#E65100')

rr_runs = [16, 25, 35, 50, 57, 77, 82, 96, 108, 118, 122, 125, 151, 161, 169, 173, 195, 217, 232, 242]
rr_wicket_overs = convert_over_format([1.3, 1.5, 4.1, 13.6, 14.2, 19.5])
rr_wicket_score = [20, 24, 50, 161, 161, 241]

plt.plot(overs, rr_runs, linestyle = '--', linewidth = 2.5, color = '#EA1A8C', label = 'RR')
plt.plot(rr_wicket_overs, rr_wicket_score, linestyle = 'None', marker = 'o', markersize = 12, markeredgewidth = 3, markeredgecolor = '#EA1A8C')

plt.xlabel('Overs')
plt.ylabel('Score')
plt.legend()
plt.tight_layout()
```
