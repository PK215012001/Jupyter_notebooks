```python
# we can even group by the columns
country_grp = df.groupby(['Country'])
country_grp
```

```python
country_grp.get_group('India')
```

```python
# let's find out social media users in india
filt = (df['Country'] == 'India')
df.loc[filt, 'SocialMedia'].value_counts()
```

```python
# we can do this like below also
country_grp.get_group('India')['SocialMedia'].value_counts()
```

```python
# we can even do like this
country_grp['SocialMedia'].value_counts().loc['India']
```

```python
# if we want to find median for each and every country
country_grp['ConvertedComp'].median()
```

```python
# If we want multiple aggregate functions
country_grp['ConvertedComp'].agg(['median', 'mean'])
```

```python
# If we want to know how many people know python in india
filt = (df['Country'] == 'India')
df.loc[filt, 'LanguageWorkedWith'].str.contains('Python').sum()
```

```python
# if we want to do it by using group, we have to use apply function
country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum()).loc['India']
```

```python
country_respondents = df['Country'].value_counts()
country_respondents
```

```python
country_uses_python = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
country_uses_python
```

```python
df2 = pd.concat([country_respondents, country_uses_python], axis = 'columns', sort = False)
df2
```

```python
df2.rename(columns={'count': 'NumRespondents', 'LanguageWorkedWith': 'NumKnowsPython'}, inplace=True)
```

```python
df2
```

```python
df2['PctKnowsPython'] = (df2['NumKnowsPython']/df2['NumRespondents']) * 100
df2
```
