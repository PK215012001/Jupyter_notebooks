```python
people = {
    'first': ['Corey', 'Jane', 'John', 'Adam'], 
    'last': ['Schafer', 'Doe', 'Doe', 'Doe'], 
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', 'A@email.com']
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(people)
df
```

```python
# sorting rows by last name
df.sort_values(by = 'last')
```

```python
df
```

```python
# In descednign order
df.sort_values(by = ['last', 'first'], ascending = False)
```

```python
# we can even sort some by ascendning and some by descending
df.sort_values(by = ['first', 'last'], ascending = [True, False], inplace  =True)
df
```

```python
# we can even sort by index
df.sort_index()
```

```python
# If we want to just check the series sorting
df['email'].sort_values()
```
