```python
# let's filter candidates with high salary and find out there salary and country and language workedwith
filter = (df['ConvertedComp'] > 70000)
df.loc[filter, ['Country', 'ConvertedComp', 'LanguageWorkedWith']]
```

```python
# Let'ss ay we want to know the candidates present in below countries
countries = ['United States', 'India', 'Canada']
filt = df['Country'].isin(countries)
df.loc[filt]
```

```python
# let's say we want to know how many people know python, na doesn't includes null values
filt = (df['LanguageWorkedWith'].str.contains('Python', na =False))
df.loc[filt]
```

```python
# Let's say we want to know how many persons with >70k salary knows python and are from india
filt = (df['LanguageWorkedWith'].str.contains('Python', na = False)) & (df['Country'] == 'India') & (df['ConvertedComp'] > 70000)
df.loc[filt].head(3)
```
